```python
# === Cell 1: Upload & Load ===
# Works for CSV, XLSX, JSON, PARQUET. File is loaded into a pandas DataFrame called `df`.

import io
import json
import sys
import pandas as pd
import numpy as np

# Colab uploader
try:
    from google.colab import files
except Exception as e:
    raise RuntimeError("This template is meant for Google Colab. Open it in Colab.") from e

uploaded = files.upload()  # Pick your dataset file

if not uploaded:
    raise ValueError("No file uploaded.")

fname = next(iter(uploaded.keys()))
raw = uploaded[fname]

def _try_read_csv(bytes_obj):
    # Robust CSV reader with delimiter sniffing
    import csv
    sample = bytes_obj[:2048].decode("utf-8", errors="ignore")
    try:
        dialect = csv.Sniffer().sniff(sample, delimiters=[",",";","|","\t"])
        sep = dialect.delimiter
    except Exception:
        sep = ","
    for enc in ("utf-8", "latin-1"):
        try:
            return pd.read_csv(io.BytesIO(bytes_obj), sep=sep, encoding=enc)
        except Exception:
            pass
    # Last resort
    return pd.read_csv(io.BytesIO(bytes_obj), sep=sep, engine="python", encoding_errors="ignore")

ext = fname.lower().rsplit(".", 1)[-1] if "." in fname else ""

if ext == "csv":
    df = _try_read_csv(raw)
elif ext in ("xls", "xlsx"):
    df = pd.read_excel(io.BytesIO(raw))
elif ext == "json":
    # Try records-or-table styles, fall back to normalizer
    try:
        df = pd.read_json(io.BytesIO(raw), orient="records")
    except Exception:
        obj = json.loads(raw.decode("utf-8", errors="ignore"))
        df = pd.json_normalize(obj)
elif ext in ("parquet", "pq"):
    df = pd.read_parquet(io.BytesIO(raw))
else:
    # Make a best effort: try CSV first, then Excel, then JSON
    try:
        df = _try_read_csv(raw)
    except Exception:
        try:
            df = pd.read_excel(io.BytesIO(raw))
        except Exception:
            try:
                obj = json.loads(raw.decode("utf-8", errors="ignore"))
                df = pd.json_normalize(obj)
            except Exception as e:
                raise ValueError(f"Could not parse file '{fname}'. "
                                 "Supported: csv, xlsx, json, parquet.") from e

print(f"Loaded file: {fname}")
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns\n")

# Quick sanity checks
with pd.option_context("display.max_rows", 8, "display.max_columns", 20):
    print("== HEAD ==")
    display(df.head())
    print("\n== INFO ==")
    df.info()
    print("\n== NULL COUNTS ==")
    display(df.isna().sum().to_frame("null_count").T)
```

```python
# === Rosalind string slice problem ===
# Upload your dataset file containing:
# line 1: string s
# line 2: four integers a b c d

from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, "r", encoding="utf-8") as f:
    lines = f.read().strip().splitlines()

s = lines[0].strip()
a, b, c, d = map(int, lines[1].split())

# inclusive slicing in Python: use b+1, d+1
part1 = s[a:b+1]
part2 = s[c:d+1]

answer = part1 + " " + part2
print(answer)

# optional save
with open("answer.txt", "w") as f:
    f.write(answer)
print("Answer saved to answer.txt")
```

```python
# === Rosalind odd sum problem ===
# Upload dataset file containing: two integers a b (with a < b < 10000)

from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, "r", encoding="utf-8") as f:
    a, b = map(int, f.read().strip().split())

# Sum odd numbers in range [a, b]
answer = sum(n for n in range(a, b+1) if n % 2 == 1)
print(answer)

# optional: save for submission
with open("answer.txt", "w") as f:
    f.write(str(answer))
print("Answer saved to answer.txt")
```

```python
# === Rosalind: Translating RNA into Protein ===
# Upload a dataset file containing a single RNA string s (≤ 10 kbp)

from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, "r", encoding="utf-8") as f:
    s = f.read().strip()

# RNA codon table
codon_table = {
    'UUU':'F','CUU':'L','AUU':'I','GUU':'V',
    'UUC':'F','CUC':'L','AUC':'I','GUC':'V',
    'UUA':'L','CUA':'L','AUA':'I','GUA':'V',
    'UUG':'L','CUG':'L','AUG':'M','GUG':'V',
    'UCU':'S','CCU':'P','ACU':'T','GCU':'A',
    'UCC':'S','CCC':'P','ACC':'T','GCC':'A',
    'UCA':'S','CCA':'P','ACA':'T','GCA':'A',
    'UCG':'S','CCG':'P','ACG':'T','GCG':'A',
    'UAU':'Y','CAU':'H','AAU':'N','GAU':'D',
    'UAC':'Y','CAC':'H','AAC':'N','GAC':'D',
    'UAA':'Stop','CAA':'Q','AAA':'K','GAA':'E',
    'UAG':'Stop','CAG':'Q','AAG':'K','GAG':'E',
    'UGU':'C','CGU':'R','AGU':'S','GGU':'G',
    'UGC':'C','CGC':'R','AGC':'S','GGC':'G',
    'UGA':'Stop','CGA':'R','AGA':'R','GGA':'G',
    'UGG':'W','CGG':'R','AGG':'R','GGG':'G'
}

# Translate RNA string into protein
protein = []
for i in range(0, len(s), 3):
    codon = s[i:i+3]
    if len(codon) < 3:  # ignore incomplete codon at end
        continue
    amino_acid = codon_table.get(codon, "")
    if amino_acid == "Stop":
        break
    protein.append(amino_acid)

protein_str = "".join(protein)
print(protein_str)

# Save to answer.txt
with open("answer.txt", "w", encoding="utf-8") as f:
    f.write(protein_str)

print("Answer saved to answer.txt")
```

```python
# === Rosalind: GC Content (FASTA) ===
# Upload a FASTA file with up to 10 DNA strings (≤ 1 kbp each).
# Output: ID with highest GC-content and its GC percentage.

from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))

def parse_fasta(path):
    records = []
    with open(path, "r", encoding="utf-8") as f:
        label = None
        chunks = []
        for line in f:
            line = line.strip()
            if not line:
                continue
            if line.startswith(">"):
                if label is not None:
                    records.append((label, "".join(chunks)))
                label = line[1:].strip()
                chunks = []
            else:
                chunks.append(line)
        if label is not None:
            records.append((label, "".join(chunks)))
    return records

def gc_content(seq: str) -> float:
    s = seq.replace(" ", "").replace("\n", "")
    s = s.upper()
    if not s:
        return 0.0
    gc = s.count("G") + s.count("C")
    return 100.0 * gc / len(s)

records = parse_fasta(filename)
if not records:
    raise ValueError("No FASTA records found.")

best_id, best_gc = None, -1.0
for rid, seq in records:
    gc = gc_content(seq)
    if gc > best_gc:
        best_gc = gc
        best_id = rid

# Print results (ID and GC percentage). Precision: 6 decimals is safe.
print(best_id)
print(f"{best_gc:.6f}")

# Save to answer.txt for easy submission
with open("answer.txt", "w", encoding="utf-8") as f:
    f.write(f"{best_id}\n{best_gc:.6f}\n")
print("Answer saved to answer.txt")
```
